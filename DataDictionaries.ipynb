## Game Consoles Dictionary

| Column Name               | Description |
|---------------------------|-------------|
| Console                   | Name of the gaming console (e.g., PlayStation 2, Nintendo DS). |
| Type                      | Device category: Home, Handheld, or Home console add‑on. |
| Company                   | Manufacturer of the console. |
| Console Release Year      | Year the console was officially released. |
| Console Disco Year        | Year the console was discontinued. |
| Consoles sold (million)   | Total units sold worldwide (millions). |
| Remarks                   | Additional notes or context. |


## Game Sales Data Dictionary

| Column Name           | Description |
|-----------------------|-------------|
| Game Title            | Name of the video game. |
| Console               | Console/platform the game was released on. |
| Year                  | Release year of the game. |
| Genre                 | Game genre (e.g., Action, Sports, RPG). |
| Company               | Publisher or developer associated with the game. |
| North America_Sales   | Sales in North America (millions). |
| Europe_Sales          | Sales in Europe (millions). |
| Japan_Sales           | Sales in Japan (millions). |
| Global_Sales          | Total worldwide sales (millions). |
