```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Steps Performed
1. Import Required Libraries
The workflow uses:
- pandas for data manipulation
- matplotlib and seaborn for visualization
- numpy for numerical operations
2. Load the Dataset
Two variations of read_csv() are shown:
- A standard load with on_bad_lines='skip' to ignore malformed rows
- A semicolon-delimited version using sep=';'
This ensures flexibility depending on the dataset’s formatting.
3. Clean the North America_Sales Column
The cleaning pipeline:
- Replaces invalid entries ("", "N/A", "NA", None) with 0
- Converts the column to float to handle mixed types
- Fills any remaining missing values with 0
- Converts the cleaned values to int for consistency
This ensures the column is fully numeric and ready for aggregation or visualization.


```python
df = pd.read_csv("../Data/Dagevolve_full.csv", on_bad_lines='skip')



df = pd.read_csv("../Data/Dagevolve_full.csv", sep=';', on_bad_lines='skip')

df["North America_Sales"] = (
    df["North America_Sales"]
    .replace(["", "N/A", "NA", None], 0)
    .astype(float)
    .fillna(0)
    .astype(int)
)

df.head()



```
