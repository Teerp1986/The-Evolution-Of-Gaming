```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = r"C:\Users\Tonika Price\Source\Repos\EvolutionOfGaming\Data\Dagevolve_full.csv"

df = pd.read_csv(
    file_path,
    sep=';',                
    on_bad_lines='skip',
    encoding='utf-8'
)


Genre = df.groupby("Genre")["Global_Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 7))


base_color = "#A2E1A4"
highlight_color = "#F1A4B6"

colors = [base_color] * len(Genre)


max_index = Genre.idxmax()
highlight_position = Genre.index.get_loc(max_index)
colors[highlight_position] = highlight_color


bars = plt.bar(Genre.index, Genre.values, color=colors, width=0.75)


for i, value in enumerate(Genre.values):
    plt.text(
        i,
        float(value) * 0.50,
        f"{value:.1f}",
        ha="center",
        va="center",
        fontsize=8,
        fontweight="bold",
        color="#1A1A1A"
    )


plt.gca().yaxis.grid(
    True,
    linestyle="--",
    linewidth=1.2,
    alpha=0.5,
    color="#A7A7A7"
)


ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_linewidth(1.1)
ax.spines["bottom"].set_linewidth(1.1)


plt.title("Global Highest Grossing Genre as of 2020", fontsize=20, fontweight="bold", pad=15)
plt.xlabel("Genre", fontsize=13, fontweight="bold")
plt.ylabel("Global Sales (Millions)", fontsize=13, fontweight="bold")

plt.xticks(rotation=40, ha="right", fontsize=11)

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV
file_path = r"C:\Users\Tonika Price\Source\Repos\EvolutionOfGaming\Data\Dagevolve_full.csv"

df = pd.read_csv(
    file_path,
    sep=';',                 
    on_bad_lines='skip',
    encoding='utf-8'
)

# Compute top 10 companies by total consoles sold
top_companies = (
    df.groupby("Game_Company")["Consoles sold (million)"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
      .reset_index()
)

# --- Plotting ---
plt.figure(figsize=(12, 6))

max_value = top_companies["Consoles sold (million)"].max()

colors = [
    "#7F00FF" if value == max_value else "#C9A9FF"
    for value in top_companies["Consoles sold (million)"]
]

sns.barplot(
    data=top_companies,
    x="Consoles sold (million)",
    y="Game_Company",
    hue="Game_Company",
    palette=colors,
    legend=False
)

# Add labels inside bars
for i, value in enumerate(top_companies["Consoles sold (million)"]):
    plt.text(
        value / 2,
        i,
        f"{value:.1f}",
        ha="center",
        va="center",
        fontsize=11,
        fontweight="bold",
        color="black"
    )

plt.title("Top 10 Console Companies by Units Sold since 1980", fontsize=16)
plt.xlabel("Units Sold (Millions)", fontsize=12)
plt.ylabel("Company", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSV
file_path = r"C:\Users\Tonika Price\Source\Repos\EvolutionOfGaming\Data\Dagevolve_full.csv"

df = pd.read_csv(
    file_path,
    sep=';',                 
    on_bad_lines='skip',
    encoding='utf-8'
)

# --- Clean + prepare data (CSV equivalent of your SQL logic) ---

# Drop missing values
df = df.dropna(subset=["Year", "Type"])

# Convert Year to integer
df["Year"] = df["Year"].astype(int)

# Assign Decade (same logic as SQL CASE WHEN)
df["Decade"] = pd.cut(
    df["Year"],
    bins=[1989, 1999, 2009, 2019, 2029],
    labels=["1990s", "2000s", "2010s", "2020s"]
)

# Remove rows where decade is NaN
df = df.dropna(subset=["Decade"])

# Group by Decade + Type and count
df_grouped = (
    df.groupby(["Decade", "Type"])
      .size()
      .reset_index(name="Type_Count")
)

# --- Pivot for plotting ---
pivot_df = df_grouped.pivot(index="Decade", columns="Type", values="Type_Count").fillna(0)

# Extract labels
decades = pivot_df.index.tolist()
types = pivot_df.columns.tolist()

# Numeric x positions
x = np.arange(len(decades))
width = 0.25  # bar width

plt.figure(figsize=(12, 6))

# Plot each console type
for i, t in enumerate(types):
    bars = plt.bar(x + i*width, pivot_df[t], width=width, label=t)

    # Add labels on bars
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            height,
            f"{int(height)}",
            ha='center',
            va='bottom',
            fontsize=10
        )

# X-axis labels
plt.xticks(x + width, decades)

# Titles and labels
plt.title("Most Popular Console Types by Decade", fontsize=16)
plt.xlabel("Decade", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Legend & grid
plt.legend(title="Console Type")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```
